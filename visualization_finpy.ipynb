### Two dimensional datasets

```python
y = np.random.standard_normal((20,2)).cumsum(axis = 0 )
```

```python
plt.figure(figsize =(10,6))
plt.plot(y, lw = 1.5)
plt.plot(y, 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
```

```python
#labelling data

plt.figure(figsize = (10,6))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc = 0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
```

```python
y[:, 0] = y[:, 0] * 100
```

```python
plt.figure(figsize = (10,6))
plt.plot(y[:, 0], lw = 1.5, label = '1st')
plt.plot(y[:, 1], lw = 1.5, label = '2nd')
plt.plot(y, 'ro')
plt.legend(loc = 0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
```

```python
# solutions to scale problem in the plot above
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw = 1.5, label = '1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc = 8)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A simple plot')
ax2 = ax1.twinx()
plt.plot(y[:, 1], 'g', lw = 1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc = 0)
plt.ylabel('value 2nd')
```

```python
#alternative solution is to make two subplots
plt.figure(figsize=(10, 6))
plt.subplot(211)  
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212)  
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value');
```

```python
# we can make totally different graph on 2nd subplot

plt.figure(figsize = (10,6))
plt.subplot(121)
plt.plot(y[:, 0], lw = 1.5, label = '1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc = 0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('1st data set')
plt.subplot(122)
plt.bar(np.arange(len(y)), abs(y[:, 1]), width = 0.5,
       color = 'g', label = '2nd')
plt.legend(loc = 0)
plt.xlabel('index')
plt.title('2nd data set')
```

### other plot types

```python
y = np.random.standard_normal((1000,2))
plt.figure(figsize = (10,6))
plt.plot(y[:, 0], y[:, 1], 'ro')
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plot')
```

```python
#specific function for gennerating scatter plot
plt.figure(figsize = (10,6))
plt.scatter(y[:, 0], y[:, 1], marker = 'o')
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plot')
```

```python
#visualising third dimension in scatterplot

c = np.random.randint(0,10, len(y))
plt.figure(figsize = (10, 6))
plt.scatter(y[:, 0], y[:, 1], c=c, cmap='coolwarm', marker = 'o')
plt.colorbar()
```

```python
# histogram

plt.figure(figsize = (10, 6))
plt.hist(y, label = ['1st', '2nd'], bins = 25)
plt.legend(loc = 0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

```python
#boxplots

fig, ax = plt.subplots(figsize = (10, 6))
plt.boxplot(y)
plt.setp(ax, xticklabels = ['1st', '2nd']) #set properties
plt.xlabel('data set')
plt.ylabel('value')
plt.title('Boxplot')
```

```python
def func(x):
    return 0.5 * np.exp(x) + 1
a, b = 0.5, 1.5
x = np.linspace(0,2)
y = func(x)
Ix = np.linspace(a,b)
Iy = func(Ix)
verts = [(a,0)] + list(zip(Ix, Iy)) + [(b,0)]
```

```python
from matplotlib.patches import Polygon

fig,ax = plt.subplots(figsize = (10,6))
plt.plot(x,y, 'b', linewidth = 2)
plt.ylim(bottom = 0)
poly = Polygon(verts, facecolor = '0.7', edgecolor = '0.5')
ax.add_patch(poly)
plt.text(0.5 * (a + b), 1, r'$\int_a^b f(x)\mathrm{d}x$',
                  horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')
ax.set_xticks((a,b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))
```
